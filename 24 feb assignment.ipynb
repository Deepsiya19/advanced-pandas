# Q5. How are DataFrame.size() and DataFrame.shape() different?

The `DataFrame.size` and `DataFrame.shape` attributes in pandas DataFrame are used to provide different types of information about the DataFrame:

1. **DataFrame.size**:
   - The `DataFrame.size` attribute returns the total number of elements in the DataFrame. It calculates the size by multiplying the number of rows by the number of columns.
   - It returns a single integer value representing the total number of elements in the DataFrame, including both data and missing values (NaNs).
   - The formula to calculate the size is: `DataFrame.size = DataFrame.shape[0] * DataFrame.shape[1]`.

   Example:
   ```python
   import pandas as pd
   df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
   print(df.size)  # Output: 6
   ```

2. **DataFrame.shape**:
   - The `DataFrame.shape` attribute returns a tuple representing the dimensions of the DataFrame. The tuple contains two elements: the number of rows and the number of columns, respectively.
   - It provides information about the structure of the DataFrame, indicating how many rows and columns it contains.
   - It does not include any information about the data within the DataFrame, such as the actual values or missing values.

   Example:
   ```python
   import pandas as pd
   df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
   print(df.shape)  # Output: (3, 2)
   ```

In summary, `DataFrame.size` provides the total number of elements in the DataFrame, while `DataFrame.shape` provides the dimensions of the DataFrame (number of rows and columns).

# Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.

````python
data = {
    'Sales': [100, 150, 200, 250, 300, 350, 400, 450, 500, 550],
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05',
             '2023-01-06', '2023-01-07', '2023-01-08', '2023-01-09', '2023-01-10']
}

df = pd.DataFrame(data)
df
````

````python
df['Moving Average']=df['Sales'].rolling(window=7,min_periods=1).mean()
df
````
